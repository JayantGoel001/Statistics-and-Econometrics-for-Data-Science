## Diagnosing Bias and Variance in a dataset(practical Implementation):

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
```

```python
dataset=pd.read_csv('..\data\\food_outlet_data.csv')
```

```python
dataset.head()
```

```python
train_data,test_data=train_test_split(dataset,test_size=0.3,random_state=47)
m=train_data['population'].values.size
```

```python
train_error=[]
test_error=[]
# training the model by increasing train dataset one by one.
for i in range(1,m+1):
    X_train=train_data.values[0:i,0].reshape(-1,1)
    Y_train=train_data.values[0:i,1].reshape(-1,1)
    model=LinearRegression().fit(X_train,Y_train)
    predict_train=model.predict(X_train)
    train_error.append(mean_squared_error(Y_train,predict_train))
    predict_test=model.predict(test_data['population'].values.reshape(-1,1))
    test_error.append(mean_squared_error(test_data['profit'],predict_test))
```

```python
no_of_set=np.arange(1,m+1,1)
train_error=np.array(train_error) # converting list into array.
test_error=np.array(test_error)
```

```python
plt.plot(no_of_set,train_error)
plt.plot(no_of_set,test_error)
plt.xlabel('No.of training set')
plt.ylabel('error')
plt.title('learning curve')
plt.legend(['train error','test error'],loc='upper right')
plt.show()
```
